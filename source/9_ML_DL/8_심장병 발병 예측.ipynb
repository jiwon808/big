```python
import pandas as pd                                # 결측치 처리하기 위해
import numpy as np                                 # numpy배열 위해 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout  # Dense를 쓰기 위해  #과적합 피하기 위해 Dropout필요 (earlystopping과 비슷)
import matplotlib.pyplot as plt                    #그래프
```

(2) 전처리 (결측치 날림) 

(3) x,y나눔.  input라고하고 target이라고 하고

(5) X측 데이터 scale 조정 (평균0 표준편차1로 조정) 

(6) 학습용데이터 , test테이터로 나눔 (검증 데이터(validation_data)가 없음 ) 

 

```python
#1. 데이터 셋 생성
raw_data = pd.read_excel('data/heart-disease.xlsx')   #excel확장자이기에 pd.read_excel()
raw_data.head()
```

```python
clean_data=raw_data.replace("?",np.nan)     #pd.numeric 숫자로 바꿔줌
clean_data
```

- **NaN값이 있는 row을 제거** 

```python
clean_data= clean_data.dropna() #NaN값이 있는 row을 제거 
clean_data
```

- **결측치 제거되었는지 확인**

```python
clean_data.info()
```

 - **null마다 결측치 있나없나**

```python
clean_data.isnull().sum()    #null마다 결측치 있나없나
```

- **데이터가 이진 분류이기에 원핫인코딩 안해도됨** 

컴럼 이름 확인  **.columns**

```python
clean_data.columns
```

```python
Input = clean_data.iloc[:,:-1]
Target = clean_data.iloc[:,[-1]]
Target
```

- **심장병 걸린 데이터 수**

```python
# 심장병 걸린 데이터 수 
Target['heartDisease'].sum()
```

- **심장병 걸린 데이터 비율**

```python
# 심장병 걸린 데이터 비율
Target['heartDisease'].mean()
```

### **scale조정**
**from sklearn.preprocessing import StandardScaler**

```python
#scale 조정
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(Input)                         # scale조정  '전'의 데이터
scaled_input = scaler.transform(Input)   # scale조정  '후'의 데이터
scaled_input
```

scaled_input을 데이터프레임으로 변환

```python
scaled_input = pd.DataFrame(scaled_input)
scaled_input
```

```python
scaled_input.describe().T   #mean은 0에 가깝고, std(표준편차)는 1에 가까움   #scale은 이에 맞춰서 알아서 조정해줌 
```

### 학습데이터: 테스트 데이터 = 7:3으로 나눔

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(scaled_input,Target,test_size=0.3,random_state=5)  
                                                #random_state은 seed값 설정과 같음. 매번 random값을 같게 출력
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)   #7:3로 나눠진 numpy배열
print(type(X_train))
```

```python
#데이터프레임을 넣어도 numpy로 바꿔짐           내가 바꾸려면 to_numpy()
```

# 모델구성

```python
#2. 모델 구성
model= Sequential()
model.add(Dense(units=500, input_dim=13, activation="tanh")) # 500x13+500 = 7000
#과적합을 해결하고자 13개 입력하고 500개 나옴. 500개 관련된 weight도 나옴. 500개 받고 또 200개 출력  (500개 기억하되 두루뭉실하게 기억하자. 그러면 과적합 방지) 
#과적합을 해결하고자 두루뭉실하게 기억하자 (Dropout사용)
model.add(Dropout(0.3))    #결과 나오면 30% 버리고
model.add(Dense(units=200,activation="tanh")) 
model.add(Dropout(0.3))
model.add(Dense(units=100, activation="tanh"))  # 200x100+100 =20100 
model.add(Dropout(0.3))
model.add(Dense(units=1, activation="sigmoid" ))  # 100x1+1 =101개 파라미터 필요 
model.summary()                                   # 127,401개의 파라미터가 학습할때 마다 업그레이드 됨 
```

# 학습과정 설정 
- compile 이용

```python
#3. 학습과정 설정
from tensorflow.keras import metrics
model.compile(loss="mse", optimizer="adam", metrics=['accuracy'])  #최소값이 되는 방향으로   #adam 방향도 사이즈도 적절하게 
# model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=[metrics.Recall()])  #recall,precision은 그냥 쓰면 안되고 이와 관련된 api를 써야 #Recall()하면 재현률 계산

# accuracy (정확도): 예측결과와 실제 값이 동일한 건수 / 전체 건수
# Recall(재현율):      실제 True인데 + 예측도 True로 한 건수 / 실제 True인 전체 건수
# Precision (정밀도):  실제True인데 + 예측도True한 건수/True로 예측한 건수   
# model.compile(loss="mse", optimizer="adam", metrics=['metrics.Precision()'])
```

# 학습시키기


```python
#4. 학습시키기
hist = model.fit(X_train, Y_train, epochs=50, batch_size=50, validation_split=0.2, verbose=2)

#5.학습과정 표시하기 
fig, loss_ax = plt.subplots()
loss_ax.plot(hist.history['loss'], 'y',label="train loss")
loss_ax.plot(hist.history['val_loss'],'g',label='val loss')   #val_loss 검증셋 데이어틔 loss   #200번이 넘어가면서 너무 과적합해서 loss가 높아져. 회귀선과 더 멀어짐
loss_ax.set_xlabel("epochs")
loss_ax.set_ylabel("loss")

acc_ax = loss_ax.twinx()  #x축을 공유하는 acc_ax
acc_ax.plot(hist.history['accuracy'],'b', label="train accuracy")
acc_ax.plot(hist.history['val_accuracy'],'r', label="val accuracy") # accuracy 저조함. 좋은 model 아님.
acc_ax.set_ylabel("accuracy")

loss_ax.legend(loc="upper left")
acc_ax.legend(loc="lower left")
```

```python
#확인용 
hist.history.keys()
#확인용 끝
```

```python
# 5. 평가 및 예측
score = model.evaluate(X_test, Y_test, verbose=0)
print('model loss:', score[0])
print('model accuracy:', score[1])
```

```python
# 6. matrix(성능평가지표)
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import f1_score
pred = model.predict(X_test)
pred = (pred>0.5)     #0.5보다 크면 True, 작으면 False  (심장별 걸릴것이다, 안걸릴 것이다,)  (가독성을 높임)
pred
print(confusion_matrix(Y_test,pred), end='\n')
print('f1_score:', f1_score(Y_test, pred))
```

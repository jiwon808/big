```python
import numpy as np                       #csv파일을 numpy배열로 읽기        
import tensorflow.keras.utils as utils  #라벨링
from tensorflow.keras.models import Sequential # 모델 형성
from tensorflow.keras.layers import Dense, Activation 
```

<Machine Learning 프로그램 방식>
- 1. 데이터 확보 및 생성
- 2. 데이터 전처리: 훈련 데이터셋, 검증 데이터셋, 시험 데이터 셋...
- 3. 모델 구성
- 4. 모델 학습과정 (예. 회귀선 출력 100번,1000번)
- 5. 모델 학습 시키기 
- 6. 모델 평가 (시험 데이터 셋으로)
- 7. 모델 사용 (입력값을 주어서 예측값을 받기)


# 1. 데이터 셋 생성

```python
dataset = np.loadtxt('data/pima-indians-diabetes.csv',delimiter=',')
dataset         #target data:맨오른쪽   inputdata: 맨오른쪽을 제외한 나머지 
```

```python
dataset.shape
```

```python
dataset[:,:8]   #0열부터8열앞까지 즉 0-7열 
```

```python
# 학습용 데이터 (훈련 데이터)
x_train = dataset[:700,:8]
y_train = dataset[:700,8]

# 테스트 데이터
x_test = dataset[700:,:8]
y_test = dataset[700:, 8]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape   #700행 8열    700행 0열
```

 - **라벨링**
- 0 -> [1,0]  
- 1-> [0,1]

# 2. 모델 구성하기

```python
model = Sequential() 
# input_dim=8 최종 출력1
model.add(Dense(12, input_dim=8, activation="relu"))         #앞에 8개 열이 input, 맨 오른쪽 열이 output
model.add(Dense(8,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

# 3.학습과정 설정

```python
model.compile(loss="binary_crossentropy", optimizer="adam",metrics=['accuracy'])
```

# 4. 학습시키기

- **epochs를 바꾸면 'restart&runall'를 해야**
- **epochs 값을 올렸는데도 accuracy가 낮으면 2단계에서 Dense()를 올려** 

```python
hist = model.fit(x_train, y_train,epochs=1500, batch_size=64)  
```

# 5. 모델 평가하기 

**accuracy 높이는 법:(1)data를 늘려야 모델이 좋아져  --> (2) epochs -->(3) dense()값 --> (4) labeling처리해서 '분류분석 (Classfication)'으로 바꿈**

```python
score = model.evaluate(x_test,y_test)
```

```python
# loss:  score[0]            accuracy:  score[1]  
```

```python
print("%s : %.2f" % (score[0],score[1]) )
```

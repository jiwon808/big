```python
x=np.arange(0,10,0.01)
```

```python
# 방법1 (2행2열 서브플롯)
#*index* position on a grid with *nrows* rows and *ncols* columns.             
#      *index* starts at 1 in the 'upper left corner' and increases to the
#      right. *index* can also be a two-tuple specifying the (*first*,
#      *last*) indices (1-based, and including *last*) of the subplot, e.g.,
#      ``fig.add_subplot(3, 1, (1, 2))`` makes a subplot that spans the
#      upper 2/3 of the figure.
        
plt.subplot(2,2,1); plt.plot(x,np.sin(x))   
plt.subplot(2,2,2); plt.plot(x,np.cos(x))
plt.subplot(2,2,3); plt.plot(x,np.tanh(x))
plt.subplot(2,2,4); plt.plot(x,np.sin(x)*np.cos(x))
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,6))   #2행2열 figsize=(10,6)
#2차원 axes를  list(axes.flat) 이용해서 중첩 for문 안씀
for i,ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x)) #i번째 아이에 x값을 넣어라 

```

```python
fig, axes = plt.subplots(1,2,figsize=(8,5))
axes[0].scatter(x,y,marker='.')
axes[0].text(2,0,'SAMPLE TEXT',style='italic')  #x좌표2, y좌표0
axes[1].scatter(x,y,marker='*',color='r')
axes[1].annotate('Sine', xytext=(2,0.75), #글자가 있을 위치 
                xy=(5,0.5),   #화살표가 가리킬 좌표 위치
                arrowprops=dict(arrowstyle='->',
                               connectionstyle='angle3'))   # connectionstyle='angle3'   곡선
plt.show()
```

- **plot.subplots_adjust()을 사용하여 서브플롯간의 간격 변경** 
- **hspace**는 위아래 여백 **wspace** 옆여백 

```python
import numpy as np
x = np.linspace(0,10,100) #0부터10까지 100개 확보
y = np.cos(x)

fig, axes = plt.subplots(2,2,figsize=(8,5))   
plt.subplots_adjust(hspace=0.4,wspace=0.3)  # graph와graph사이 공간 늘리기 (라벨이 보이기 위해 )
                                            # hspace는 위아래 여백 wspace 옆여백 

#1번째 그래프 
axes[0,0].scatter(x,y,marker='.')
axes[0,0].set(title="An Example Axes",ylabel="y-axis", xlabel="x축", 
              xlim=[0,12], # x축 변경 
              ylim=[-2,2] )
#2번째 그래프
axes[0,1].scatter(x,y,marker='*', c='b')
axes[0,1].set_xlim(2,6)
axes[0,1].set_xlabel('x축 0~5')
axes[0,1].set_ylim(-1.5,1.5)
axes[0,1].set_ylabel('y축 -1.5~1.5')

#3번째 그래프 
axes[1,0].scatter(x,y,marker='^',c='c')
# x축 눈금위치 
axes[1,0].set_xticks(range(0,8,2))     #0부터 10까지 2씩 증가     #4개
axes[1,0].set_xticklabels([-100,3,'6개','foo'] ) #눈끔의 내용     #그러면 여기도 4개. 숫자 맞춰야. 왜냐하면 위 내용을 변경하기 때문에 
axes[1,0].set_yticks([-2,0,1,2])
axes[1,0].set_yticklabels(['-2개','제로',1,'Max'])

axes[1,1].scatter(x,y,c='r')
axes[1,1].set(xticks=[0, 2.5, 5, 7.5,10],
             xticklabels=['Min', 2.5, 5, 7.5, 'Max'],
             title='test')
axes[1,1].grid(True)
axes[1,1].spines["top"].set_visible(False)
axes[1,1].spines["bottom"].set_position(("outward",10))
plt.show()
```

## 2.6 5) 축 공유 6)그래프제목과 축제목

```python
import numpy as np
x = np.arange(0,10)
y1 = 0.5 * x**2   #0.5*(x**2)
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots()
#plt.suptitle("Main title")
ax1.plot(x,y1,'g^:')
ax1.set(xlabel='X data',ylabel="y1 data")

ax2 = ax1.twinx() # x축을 공유하는 ax2
ax2.plot(x,y2,'b--')       #x축을 공유하고 y축은 따로 
ax2.set_ylabel('y2 data', color='b')

ax3 = ax1.twiny() # ax1과 x축을 공유하는 ax3
ax3.plot(-x,y1,'ro-')
ax3.set_xlabel('-x data',color='r')

plt.show()
```

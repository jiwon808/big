### 간략한 문법
- \d (숫자와 매치.[0-9] 동일)
- \D (숫자가 아닌 것)
- \s (whitespace)
- \w (영문자나 숫자)
- \W (영문자나 숫자가 아닌 문자)
- . (문자 하나)
- plus 기호  " +   " (1번 이상 반복)
- 별표 기호 " * " (0번 이상 반복)
- {2,4} (2~4회 반복)
- ? ({0,1}을 의미)

**예제) 전국 날씨 RSS 데이터 가져오기**

- **<정규표현식을 활용>** <br>
**decode('utf-8')**를 해야 **'한글'**이 나옴 

```python
import re, urllib.request
url="http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = urllib.request.urlopen(url)
contents = target.read().decode('utf-8')   #   decode('utf-8')를 해야 '한글'이 나옴 
#content
locations= re.findall(r'<location wl_ver="3">.+?</location>',contents, re.DOTALL)    
                                                                        #re.DOTALL은 정규표현식의 .이 줄바꿈 문자를 포함
#len(locations)
for loc in locations:
    province = re.search(r'<province>(.+)</province>',loc)         #실제 데이터안 <province>서울ㆍ인천ㆍ경기도</province>
    city = re.search(r'<city>(.+)</city>',loc)                                  # <city>파주</city>
    print(province.group(1),'지역의',city.group(1))
    data = re.findall(r'<data>.+?</data>',loc,re.DOTALL)    # .+는 한글자 이상 이라는 뜻 
    for item in data:
        tmEf = re.search(r'<tmEf>(.+)</tmEf>',item).group(1)
        wf = re.search(r'<wf>(.+)</wf>',item).group(1)
        tmn= re.search(r'<tmn>(.+)</tmn>',item).group(1)
        tmx=re.search(r'<tmx>(.+)</tmx>',item).group(1)
        print("\t날짜시간:{}, 날씨:{},최저:{}도,최고:{}도".format(tmEf,wf,tmn,tmx))
```

```python
#select locations 
# select all
# tmEf  select_one 

import urllib.request
from bs4 import BeautifulSoup 
url="http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = urllib.request.urlopen(url)
weather=requests.get(url)
soup= BeautifulSoup(weather.content,"html.parser")   # html.parser를 하면 전부다 '소문자'로 바꿈 
locations = soup.select("location")
for loc in locations:
        province = loc.select_one("province").text  #하나 찾음   #location안에 province가 하나만 있어서 select_one
                                                    # 혹은 province = loc.select("province")[0].text  *************************
        city = loc.select_one("city").string   #.string은 .text과 같음  ****
        print(province,'지역의',city)
        # 여기서부터 2단계 
        data = loc.select('data')   #실제 데이터 안에 <data> </data>안에 있는 내용을 전부 가져옴 
        ampm="오전"
        for item in data:
            tmEf = item.select_one('tmEf').text[:10]  #10번째 앞까지 (0번방부터시작. 따라서 0번부터9번까지 출력. 여기서 '날짜')  
            #parser하면 소문자로 내용이 출력.그러나 html이어서 select_one()안에 대소문자 상관없이 쳐도됨  ********
            wf = item.select_one('wf').text
            tmn = item.select_one('tmn').text
            tmx =  item.select_one('tmx').text
            print("\t{}{}, 날씨:{},기온:{}도~{}도".format(tmEf,ampm,wf,tmn,tmx))
            
            if ampm=="오전":     # 실제 데이터에  오전한번 오후 한번   --> 따라서  만약에 오전한번 출력했으면, 오후출력해라. 오후 출력했으면 오전 출력해라 
                ampm = "오후"      #************************************************
            else: 
                ampm="오전" 


```
